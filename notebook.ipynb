```python
def get_phase(over):
    if over <= 6:
        return "Powerplay"
    elif over <= 15:
        return "Middle Overs"
    else:
        return "Death Overs"

df['phase'] = df['over'].apply(get_phase)
```

```python
df['dot_ball'] = df['runs_total'].apply(lambda x: 1 if x == 0 else 0)
```

```python
df['boundary'] = df['runs_batter'].apply(lambda x: 1 if x in [4,6] else 0)
```

```python
df['is_wicket'] = df['wicket_kind'].notnull().astype(int)
```

```python
df['single'] = df['runs_batter'].apply(lambda x: 1 if x == 1 else 0)

df['double'] = df['runs_batter'].apply(lambda x: 1 if x == 2 else 0)
```

```python
df['pressure_ball'] = (
    (df['dot_ball'] == 1) | (df['is_wicket'] == 1)
).astype(int)
```

Toss impact based on IPL seasons

```python
def clean_season(season):

    season = str(season)

    # Handle formats like 2007/08
    if '/' in season:

        parts = season.split('/')

        # Example:
        # 2007/08 -> 2008
        # 2020/21 -> 2021

        last_part = parts[-1]

        if len(last_part) == 2:
            return int('20' + last_part)

        return int(last_part)

    return int(season)

# Apply cleaning
df['season'] = df['season'].apply(clean_season)
```

```python
season_toss = (
    df[df['toss_winner'] == df['winner']]
    .groupby('season')['match_id']
    .nunique()
)

season_total = (
    df.groupby('season')['match_id']
    .nunique()
)

season_toss_percentage = (
    (season_toss / season_total) * 100
).reset_index()

season_toss_percentage.columns = [
    'Season',
    'Toss_Win_Percentage'
]

print(season_toss_percentage)
```

```python
season_batters = (
    df.groupby(['season', 'batter'])['runs_batter']
    .sum()
    .reset_index()
)

print(season_batters.head())
```

```python


# Rank batters within each season
season_batters['rank'] = (
    season_batters
    .groupby('season')['runs_batter']
    .rank(
        method='first',
        ascending=False
    )
)

# Keep top 5
top5_batters = season_batters[
    season_batters['rank'] <= 5
]

# Sort properly
top5_batters = top5_batters.sort_values(
    ['season', 'rank']
)

# Clean columns
top5_batters = top5_batters[
    ['season', 'rank', 'batter', 'runs_batter']
]

# Rename columns
top5_batters.columns = [
    'Season',
    'Rank',
    'Batter',
    'Runs'
]

print(top5_batters)
```

```python
# Batter statistics by season
season_stats = (
    df.groupby(['season', 'batter'])
    .agg({

        'runs_batter':'sum',

        'ball':'count',

        'match_id':'nunique',

        'wicket_player_out':'count'
    })
    .reset_index()
)

# Rename columns
season_stats.columns = [

    'season',

    'batter',

    'runs',

    'balls',

    'matches',

    'dismissals'
]

# Strike Rate
season_stats['strike_rate'] = (
    (season_stats['runs'] / season_stats['balls']) * 100
).round(1)

# Batting Average
season_stats['average'] = (
    season_stats['runs'] /
    season_stats['dismissals']
).round(1)

# Handle divide-by-zero
season_stats['average'] = (
    season_stats['average']
    .fillna(season_stats['runs'])
)

# Rank by runs
season_stats['rank'] = (
    season_stats
    .groupby('season')['runs']
    .rank(
        method='first',
        ascending=False
    )
)

# Top 5
top5 = season_stats[
    season_stats['rank'] <= 5
]

# Latest season
latest_season = top5['season'].max()

latest_data = top5[
    top5['season'] == latest_season
]

latest_data = latest_data.sort_values(
    by='rank'
)
```

```python
# Total legal deliveries
legal_balls = df[
    df['extras_wides'] == 0
]

# Valid bowler wickets only
bowler_wickets = df[

    df['wicket_kind'].isin([

        'bowled',

        'caught',

        'caught and bowled',

        'lbw',

        'stumped',

        'hit wicket'
    ])
]

# Wickets per season
wickets = (
    bowler_wickets
    .groupby(['season', 'bowler'])
    .size()
    .reset_index(name='wickets')
)

# Bowler base stats
bowler_stats = (
    legal_balls.groupby(['season', 'bowler'])
    .agg({

        'runs_total':'sum',

        'ball':'count',

        'match_id':'nunique'
    })
    .reset_index()
)

# Rename columns
bowler_stats.columns = [

    'season',

    'bowler',

    'runs_conceded',

    'balls',

    'matches'
]

# Merge wickets
bowler_stats = bowler_stats.merge(

    wickets,

    on=['season', 'bowler'],

    how='left'
)

# Fill missing wickets
bowler_stats['wickets'] = (
    bowler_stats['wickets']
    .fillna(0)
    .astype(int)
)

# Overs bowled
bowler_stats['overs'] = (
    bowler_stats['balls'] / 6
).round(1)

# Economy
bowler_stats['economy'] = (
    bowler_stats['runs_conceded']
    /
    bowler_stats['overs']
).round(2)

# Sort by wickets first, then economy
bowler_stats = bowler_stats.sort_values(

    ['season', 'wickets', 'economy'],

    ascending=[True, False, True]
)

# Rank bowlers
bowler_stats['rank'] = (
    bowler_stats
    .groupby('season')
    .cumcount()
    + 1
)

# Keep top 5
top5_bowlers = bowler_stats[
    bowler_stats['rank'] <= 5
]

# Latest season
latest_season = top5_bowlers['season'].max()

latest_bowlers = top5_bowlers[
    top5_bowlers['season'] == latest_season
]

# Sort properly
latest_bowlers = latest_bowlers.sort_values(
    by='rank'
)
```
